```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-5,5,100)
```

```python
y = torch.sin(x)
```

```python
plt.plot(x.numpy(),y.numpy())
```

```python
class RNN(nn.Module):
    def __init__(self):
        super(RNN, self).__init__()
        
        self.rnn = nn.RNN(
            input_size = 1,
            hidden_size = 32,
            num_layers = 1,
            batch_first = True
        )
        self.out = nn.Linear(32, 1)
        
    def forward(self, x, h_state):
        r_out, h_state = self.rnn(x,h_state)
        return self.out(r_out), h_state
```

```python
rnn = RNN()
rnn
```

```python
loss_func = nn.MSELoss()
optimizer = torch.optim.Adam(rnn.parameters(),lr = 0.3)
```

```python
h_state = None
for epoch in range(300):
    input = x.unsqueeze(dim = 1).unsqueeze(dim = 0)
    output,h_state = rnn(input,h_state)
    loss = loss_func(output, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print('loss ', loss.data.numpy())
```

```python
output = rnn(x.unsqueeze(dim = 1).unsqueeze(dim = 0))
```

```python
plt.plot(x.numpy(),output.view(100).data.numpy())
```
